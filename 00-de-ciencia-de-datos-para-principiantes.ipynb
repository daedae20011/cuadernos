### ALCANCE
Lo que necesitamos saber sobre el alcance:
* global: cuerpo principal definido en script
* local: definido en una función
* alcance integrado: nombres en módulo de alcance integrado predefinido como print, len
<br><br>Vamos a hacer algunos ejemplos básicos

```python
# supongo que imprime qué
x = 2
def f():
    x = 3
    return x
print(x)      # x = 2 alcance global
print(f())    # x = 3 alcance global
```

```python
# ¿Qué pasa si no hay alcance local?
x = 5
def f():
    y = 2*x        # no hay alcance local x
    return y
print(f())        # usa alcance global x
# Primero se buscó en el ámbito local, luego se buscó en el ámbito global, si no se pueden encontrar dos de ellos, por último se buscó en el ámbito integrado.
```

### FUNCIÓN ANIDADA
* función dentro de la función.
* Hay una regla de LEGB que es alcance local de búsqueda, función adjunta, alcance global e integrado, respectivamente.

```python
#función anidada
def square():
    """ return square of value """
    def add():
        """ add two local variable """
        x = 2
        y = 3
        z = x + y
        return z
    return add()**2
print(square())    
```

### ARGUMENTOS POR DEFECTO y FLEXIBLES
* Ejemplo de argumento predeterminado:
<br> def f(a, b=1):
         """ b = 1 es el argumento predeterminado"""
* Ejemplo de argumento flexible:
<br> def f(*argumentos):
        """ *los argumentos pueden ser uno o más"""
<br>def f(** kwargs)
        """ **kwargs es un diccionario"""
       
<br><br> escribamos algo de código para practicar

```python
# argumentos predeterminados
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))
# ¿Qué sucede si queremos cambiar los argumentos predeterminados?
print(f(5,4,3))
```

```python
# argumentos flexibles *args
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)
# argumentos flexibles **kwargs eso es diccionario
def f(**kwargs):
    """ print key and value of dictionary"""
    for key, value in kwargs.items():               # Si no entiende esta parte, gire for loop y mire el diccionario en for loop
        print(key, " ", value)
f(country = 'spain', capital = 'madrid', population = 123456)
```

### COMPRENSIÓN DE LISTAS
**Uno de los temas más importantes de este kernel**
<br>A menudo utilizamos la comprensión de listas para el análisis de datos.
<br> comprensión de listas: colapsar bucles for para construir listas en una sola línea
<br>Ej: num1 = [1,2,3] y queremos que sea num2 = [2,3,4]. Esto se puede hacer con bucle for. Sin embargo, es innecesariamente largo. Podemos convertirlo en código de una línea que sea comprensión de listas.

```python
# Ejemplo de comprensión de lista
num1 = [1,2,3]
num2 = [i + 1 for i in num1 ]
print(num2)
```

[i + 1 para i en num1]: lista de comprensión
<br> i +1: lista de sintaxis de comprensión
<br> para i en num1: sintaxis de bucle for
<br> i: iterador
<br> num1: objeto iterable

```python
# Condicionales en iterable
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```

### ANÁLISIS EXPLORATORIO DE DATOS
value_counts(): recuentos de frecuencia
<br>valores atípicos: el valor que es considerablemente más alto o más bajo que el resto de los datos
* Digamos que el valor al 75% es Q3 y el valor al 25% es Q1.
* Los valores atípicos son más pequeños que Q1 - 1,5 (Q3-Q1) y más grandes que Q3 + 1,5 (Q3-Q1). (Q3-Q1) = RIQ
<br>Usaremos el método describe(). Describir el método incluye:
* recuento: número de entradas
* media: promedio de entradas
* std: desviación estándar
* min: entrada mínima
* 25%: primer cuantil
* 50%: mediana o segundo cuantil
* 75%: tercer cuantil
* max: entrada máxima

<br> ¿Qué es cuantil?

* 1,4,5,6,8,9,11,12,13,14,15,16,17
* La mediana es el número que está en **medio** de la secuencia. En este caso serían 11.

* El cuartil inferior es la mediana entre el número más pequeño y la mediana, es decir, entre 1 y 11, que es 6.
* El cuartil superior, encuentras la mediana entre la mediana y el número más grande, es decir, entre 11 y 17, que será 14 según la pregunta anterior.

### SERIES TEMPORALES DE PANDAS DE REMUESTRAS
* Remuestreo: método estadístico en diferentes intervalos de tiempo
     * Necesita una cadena para especificar la frecuencia como "M" = mes o "A" = año
* Reducción de muestreo: reduzca las filas de fecha y hora a una frecuencia más lenta, como de diario a semanal
* Sobremuestreo: aumente las filas de fecha y hora a una frecuencia más rápida, como de diario a cada hora
* Interpolar: Interpolar valores según diferentes métodos como 'lineal', 'tiempo' o índice'
     * https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.interpolate.html


```python
# Usaremos data2 que creamos en la parte anterior
data2.resample("A").mean()
```

```python
# Vamos a volver a muestrear con el mes
data2.resample("M").mean()
# Como puede ver, hay muchos nan porque data2 no incluye todos los meses
```

```python
# En la vida real (los datos son reales. No creados por nosotros como data2) podemos resolver este problema con interpolación
# Podemos interpolar desde el primer valor
data2.resample("M").first().interpolate("linear")
```

```python
# O podemos interpolar con mean()
data2.resample("M").mean().interpolate("linear")
```
